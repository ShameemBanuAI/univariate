```python
def get_pdf_probability(dataset,startRange,endRange):
    # plotting gragh
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startRange,color='Red')
    pyplot.axvline(endRange,color='Red')
    # generating sample
    sample=dataset
    # finding mean() and std()
    sample_mean=sample.mean()
    sample_std=sample.std()
    print("Mean = %.3f" % sample_mean)
    print("Standard deviation = %.3f" % sample_std)
    # finding distribution
    dist=norm(sample_mean,sample_std)
    # finding probabilities for the range of outcomes
    values=[value for value in range(startRange,endRange)]
    probabilities=[dist.pdf(value) for value in values ]
    prob=sum(probabilities)
    print("Area between the ({},{}) is {}". format(startRange,endRange,prob))
    return prob 
    
           
    
```

```python
get_pdf_probability(dataset["ssc_p"],60,70)
```

```python
#converting normal distribution to standard normal distribution
def stdND(dataset):
    sam_mean=dataset.mean()
    sam_std=dataset.std()
    values=[i for i in dataset ]
    z_score=[ ((j-sam_mean) / sam_std) for j in values ]
    sns.distplot(z_score,kde=True,kde_kws={'color':'black'},color='yellow')
    avg=sum(z_score)/len(z_score) 
    return avg
```
