```python
def quanqual(dataset):

    quan=[]
    qual=[]
    for colName in dataset.columns:
        if (dataset[colName].dtypes)=='O':
            qual.append(colName)
        else:
            quan.append(colName)
    return qual,quan
```

```python
qual,quan=quanqual(dataset)
```

# Central tendency

```python
for colName in quan:
    descriptive[colName]['mean']=dataset[colName].mean()
    descriptive[colName]['median']=dataset[colName].median()
    descriptive[colName]['mode']=dataset[colName].mode()[0]
```

```python
descriptive=pd.DataFrame(index=["mean","median","mode","Q1:25","Q2:50","Q3:75","99th","Q4:100"],columns=quan)
for colName in quan:
    descriptive[colName]['mean']=dataset[colName].mean()
    descriptive[colName]['median']=dataset[colName].median()
    descriptive[colName]['mode']=dataset[colName].mode()[0]
    descriptive[colName]['Q1:25']=dataset.describe()[colName]["25%"]
    descriptive[colName]['Q2:50']=dataset.describe()[colName]["50%"]
    descriptive[colName]['Q3:75']=dataset.describe()[colName]["75%"]
    descriptive[colName]['99th']=np.percentile(dataset[colName],99)
    descriptive[colName]['Q4:100']=dataset.describe()[colName]["max"]
```

# IQR (InterQuantileRange)

```python
descriptive=pd.DataFrame(index=["mean","median","mode","Q1:25","Q2:50","Q3:75","99th","Q4:100",
                               "IQR","1.5 IQR","Lesser Range","Greater Range","MinValue","MaxValue"],columns=quan)
for colName in quan:
    descriptive[colName]["mean"]=dataset[colName].mean()
    descriptive[colName]['median']=dataset[colName].median()
    descriptive[colName]['mode']=dataset[colName].mode()[0]
    descriptive[colName]['Q1:25']=dataset.describe()[colName]["25%"]
    descriptive[colName]['Q2:50']=dataset.describe()[colName]["50%"]
    descriptive[colName]['Q3:75']=dataset.describe()[colName]["75%"]
    descriptive[colName]['99th']=np.percentile(dataset[colName],99)
    descriptive[colName]['Q4:100']=dataset.describe()[colName]["max"]
    descriptive[colName]['IQR']=descriptive[colName]['Q3:75']- descriptive[colName]['Q1:25']
    descriptive[colName]['1.5 IQR']=1.5 * descriptive[colName]['IQR']
    descriptive[colName]['Lesser Range']= descriptive[colName]['Q1:25'] - descriptive[colName]['1.5 IQR']
    descriptive[colName]['Greater Range']= descriptive[colName]['Q3:75'] - descriptive[colName]['1.5 IQR']
    descriptive[colName]["MinValue"]=dataset[colName].min()
    descriptive[colName]["MaxValue"]=dataset[colName].max()
```

```python
descriptive
```

# Detection of outliers

```python
lesser=[]
greater=[]
for colName in quan:
    if (descriptive[colName]["MinValue"]) < (descriptive[colName]["Lesser Range"]):
        lesser.append(colName)
    if (descriptive[colName]["MaxValue"]) > (descriptive[colName]["Greater Range"]):
        greater.append(colName)
```

```python
lesser
```

```python
greater
```

# Replacing Outliers

```python
for lesscolName in lesser:
    dataset[lesscolName][dataset[lesscolName] < descriptive[lesscolName]["Lesser Range"]] = descriptive[lesscolName]["Lesser Range"]
    
for greatcolName in greater:
    dataset[greatcolName][dataset[greatcolName] > descriptive[greatcolName]["Greater Range"]] = descriptive[greatcolName]["Greater Range"]
```

```python
descriptive=pd.DataFrame(index=["mean","median","mode","Q1:25","Q2:50","Q3:75","99th","Q4:100",
                               "IQR","1.5 IQR","Lesser Range","Greater Range","MinValue","MaxValue"],columns=quan)
for colName in quan:
    descriptive[colName]["mean"]=dataset[colName].mean()
    descriptive[colName]['median']=dataset[colName].median()
    descriptive[colName]['mode']=dataset[colName].mode()[0]
    descriptive[colName]['Q1:25']=dataset.describe()[colName]["25%"]
    descriptive[colName]['Q2:50']=dataset.describe()[colName]["50%"]
    descriptive[colName]['Q3:75']=dataset.describe()[colName]["75%"]
    descriptive[colName]['99th']=np.percentile(dataset[colName],99)
    descriptive[colName]['Q4:100']=dataset.describe()[colName]["max"]
    descriptive[colName]['IQR']=descriptive[colName]['Q3:75']- descriptive[colName]['Q1:25']
    descriptive[colName]['1.5 IQR']=1.5 * descriptive[colName]['IQR']
    descriptive[colName]['Lesser Range']= descriptive[colName]['Q1:25'] - descriptive[colName]['1.5 IQR']
    descriptive[colName]['Greater Range']= descriptive[colName]['Q3:75'] - descriptive[colName]['1.5 IQR']
    descriptive[colName]["MinValue"]=dataset[colName].min()
    descriptive[colName]["MaxValue"]=dataset[colName].max()
```

```python
def freqtable(colName,dataset):
    freqTable=pd.DataFrame(columns=["UniqueValues","Frequency","RelativeFrequency","Cusum"])
    freqTable["UniqueValues"]=dataset[columnName].value_counts().index
    freqTable["Frequency"]=dataset[columnName].value_counts().values
    freqTable["RelativeFrequency"]=freqTable["Frequency"]/17
    freqTable["Cusum"]=freqTable["RelativeFrequency"].cumsum()
    return freqtable
```
